```python
my_data =  [10179,44985338,44442065,531815]
labels = ['active','positive','cured','death']
my_explode = (0,0,0,1.5)
my_colors = ['lightblue','lightsteelblue','pink','red']
fig1,ax1 = plt.subplots(figsize=(13,8))
plt.pie(my_data,labels=labels,autopct='%0.2f%%',startangle=15,shadow=True,colors=my_colors,explode=my_explode)
plt.axis('equal')
plt.show()
```

```python
my_data =  [9092,44986203,44444013,531818]
labels = ['New active','New positive','New cured','New death']
my_explode = (0,0,0,1.5)
my_colors = ['lightblue','lightsteelblue','pink','red']
fig1,ax1 = plt.subplots(figsize=(13,8))
plt.pie(my_data,labels=labels,autopct='%0.2f%%',startangle=15,shadow=True,colors=my_colors,explode=my_explode)
plt.axis('equal')
plt.show()
```
